```python
import nearest
import matplotlib.pyplot as plt
import math

fig, ax = plt.subplots()

# ego lane center line
poly_x = [-30,5,20,100]
poly_y = [0,0,3,3]
plt.plot(poly_x,poly_y,linestyle="--")
nearest.draw_lane(poly_x,poly_y)

# sensor objects' x and y positions
all_x, all_y, yaw = nearest.create_object_poses(80)      
nearest.draw_all_boxes(all_x, all_y, yaw)

#visu
nearest.draw_ego(ax)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.xlim(-10, 50)
plt.ylim(-10, 15)
plt.show()
```

ego lane center polyline positions are **poly_x,poly_y**

ego lan width: **5m**

object center points positions are **all_x, all_y**

ego vehicle center is (0,0)
